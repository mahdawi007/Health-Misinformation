```python

import tweepy
from datetime import datetime,timedelta
import csv
from pandas import DataFrame

# fill your own keys
access_token = ""
access_token_secret = ""
consumer_key = ""
consumer_secret = ""

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)

data = [] # empty list to which tweet_details obj will be added
def stream_tweets(dict):
    csvFile = open('CAD_results.csv', 'a')
    csvWriter = csv.writer(csvFile)
    counter = 0 # counter to keep track of each iteration
#search_terms = 'heart diseases'
    for keyword in dict:
        print ("[+] Checking tweets for keyword:", keyword)
        for status in tweepy.Cursor(api.search,
                       q=str(keyword) + " -filter:retweets",
                       since='2021-06-16', until='2021-06-24', 
                       count=1000,
                       result_type='recent',
                       include_entities=True,
                       monitor_rate_limit=True, 
                       wait_on_rate_limit=True,
                       lang="en").items():

            eastern_time = status.created_at - timedelta(5)
            edt_time = eastern_time.strftime('%Y-%m-%d %H:%M')
            csvWriter.writerow([keyword,status.created_at, status.user.screen_name.encode('utf8'), status.text.encode('utf-8')])
            tweet_details = {}
            tweet_details['keyword'] = keyword
            tweet_details['name'] = status.user.screen_name
            tweet_details['tweet'] = status.text
            tweet_details['retweets'] = status.retweet_count
            tweet_details['location'] = status.user.location
            tweet_details['created'] = status.created_at.strftime("%d-%b-%Y")
            tweet_details['followers'] = status.user.followers_count
            tweet_details['is_user_verified'] = status.user.verified
            data.append(tweet_details)
            counter += 1
            if counter % 1000 ==0:
                print(counter)
    csvFile.close()
    print('done!')

stream_tweets(dict)
df = DataFrame (data)
print(df.shape)
df.drop_duplicates(subset ="tweet",keep = False, inplace = True)
print(df.shape)
df.head(10)

```
